```python
import networkx as nx, osmnx_dir as osmnx
%matplotlib inline
osmnx.init(file_log=True, use_cache=True)
```

```python
location_point = (33.299896, -111.831638)
G = osmnx.graph_from_address('420 North Corsica Place, Chandler, Arizona', 
                             distance=800, network_type='motor', simplify=False)
```

```python
nc = ['b' if osmnx.is_endpoint(G, node) else 'r' for node in G.nodes()]
fig, ax = osmnx.plot_graph(G, node_color=nc, node_zorder=3)
```

```python
G2 = osmnx.simplify_graph(G, strict=True)
```

```python
#nc = ['b' if node in [origin, destination] else 'r' for node in G2.nodes()]
fig, ax = osmnx.plot_graph(G2, node_color='b', node_zorder=3)
```

```python
origpt = (33.307792, -111.894940)
destpt = (33.312994, -111.894998)
origin = osmnx.get_nearest_node(G2, origpt)
destination = osmnx.get_nearest_node(G2, destpt)
route = nx.shortest_path(G2, origin, destination)
```

```python
fig, ax = osmnx.plot_graph_route(G2, route)
```
