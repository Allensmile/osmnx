```python
import osmnx, networkx as nx
%matplotlib inline
```

```python
osmnx.init(file_log=True)
```

```python
G = osmnx.graph_from_place('Piedmont, California, USA')
fig, ax = osmnx.plot_graph(G, save=True, filename='place.jpg')
```

```python
# project the graph to an appropriate UTM (automatically determined) and plot it
G_projected = osmnx.project_graph(G)
fig, ax = osmnx.plot_graph(G_projected, x='x', y='y', save=True, filename='place_utm.jpg')
```

```python
# save graph to disk
osmnx.save_graph(G)
```

```python
places = ['Piedmont, California, USA', 
          'Berkeley, California, USA',
          'Emeryville, California, USA']

# use retain_all if you want to keep all disconnected subgraphs (e.g. when your places aren't adjacent)
G = osmnx.graph_from_place(places, retain_all=True)
fig, ax = osmnx.plot_graph(G, save=True, filename='places.jpg')
```

```python
G = osmnx.graph_from_place('San Francisco, California, USA')
G_projected = osmnx.project_graph(G)
fig, ax = osmnx.plot_graph(G_projected, save=True, filename='sf.jpg')
```

```python
# save graph to disk
osmnx.save_graph(G, filename='sf-projected')
```
